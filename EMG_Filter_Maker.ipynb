```python
from scipy.signal import firwin
import numpy as np

fs     = 1000   # sampling rate Hz
cutoff = 20     # high pass cutoff — removes motion artifact/DC below 20Hz
#cutoff = 500     # low pass cutoff — removes motion artifact/DC below 500Hz
n_taps = 31

# Note: pass_zero=False makes it a high pass instead of low pass
coeffs = firwin(n_taps, cutoff / (fs / 2), window='hamming', pass_zero=False)

print("const data_t hpf_coeffs[N_TAPS] = {")
for i, c in enumerate(coeffs):
    print(f"    {c:.4f}{',' if i < n_taps-1 else ''}")
print("};")
```

```python
from scipy.signal import firwin
import numpy as np

fs       = 100  # sampling rate Hz
notch_hz = 60    # frequency to remove
width    = 10    # notch width in Hz (removes 55Hz-65Hz)
n_taps   = 31

# Band stop filter — blocks a band instead of passing it
coeffs = firwin(n_taps,
                [(notch_hz - width/2) / (fs/2),
                 (notch_hz + width/2) / (fs/2)],
                window='hamming',
                pass_zero=True)  # pass_zero=True makes it a band stop

print("const data_t notch_coeffs[N_TAPS] = {")
for i, c in enumerate(coeffs):
    print(f"    {c:.4f}{',' if i < n_taps-1 else ''}")
print("};")
```
